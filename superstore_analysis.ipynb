```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/Pradeep/Desktop/RetailSalesAnalysis/Sample - Superstore.csv", encoding='latin1')
```

```python
### Total sales&profit
```

```python
total_sales= df['Sales'].sum()
total_profit= df['Profit'].sum()

print(f"Total Sales: ₹{total_sales:,.2f}")
print(f"Total Profit: ₹{total_profit:,.2f}")
```

```python
### Profit by category
```

```python
category_profit= df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
category_profit
```

```python
### Sales bu segment
```

```python
segment_sales= df.groupby('Segment')['Sales'].sum().sort_values(ascending= False)
segment_sales
```

```python
### Visualization
```

```python
plt.style.use('ggplot')

category_profit.plot(kind='bar', color='mediumseagreen',figsize=(6,4))
plt.title('Total Profit by Product Category')
plt.xlabel('Category')
plt.ylabel('Profit (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
segment_sales.plot(kind='bar', color='royalblue', figsize=(6, 4))
plt.title('Sales by Customer Segment')
plt.xlabel('Segment')
plt.ylabel('Sales (₹)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
top_subcategories = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)

top_subcategories.plot(kind='barh', color='coral', figsize=(8, 5))
plt.title("Top 10 Sub-Categories by Sales")
plt.xlabel("Total Sales (₹)")
plt.ylabel("Sub-Category")
plt.tight_layout()
plt.show()
```

```python
bottom_states = df.groupby('State')['Profit'].sum().sort_values().head(10)

bottom_states.plot(kind='barh', color='tomato', figsize=(8, 5))
plt.title("Bottom 10 States by Profit (Losses)")
plt.xlabel("Profit (₹)")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```
